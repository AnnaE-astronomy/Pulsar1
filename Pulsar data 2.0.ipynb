```python
from astropy.table import Table,vstack
import pandas
import glob
# Goes through all the components in the racs_catv tables which hold the detected objects in the mosaics
#and creates a table out of them all
for type in ['islands','components']:
 
    selavy_files=sorted(glob.glob('tables/racs_catv/*{0}.txt'.format(type)))
    data=[]
    for i,file in enumerate(selavy_files):      
        table = Table.from_pandas(pandas.read_fwf(file, skiprows=[1,]))
        
                
        table.add_column(Column(np.array(file),name="fileName"))
        data.append(table)
        
        print('Read {0} lines from {1} ({2}/{3}) \r'.format(len(data[-1]),
                                                          file,i,len(selavy_files)))
        
    racs_data=vstack(data)
    # these are empty and confuse HDF5
    del racs_data['#']
    del racs_data['comment']
    racs_data.write('RACS_StokesV_{0}.hdf5'.format(type),overwrite=True)
    print('Wrote to RACS_StokesV_{0}.hdf5'.format(type))
```

```python
#from professor Kaplan 
import psrqpy as ps
from astropy.table import Column,Table,vstack
import numpy as np
from astropy.coordinates import SkyCoord
from astropy import units as u, constants as c
 
import pickle
 
import matplotlib.pyplot as plt
max_position_uncertainty= 1*u.arcsec
max_separation= 5*u.arcsec 

columns=['Name',
          'RAJ',
          'DECJ',
          'RAJD',
          'DECJD',
          'DM',
          'S400',
          'S1400',
          'S2000',
          'ASSOC',
          'DIST',
          #'L400',
          #'L1400',
          ]


requery=True
 
if requery:
    query = ps.QueryATNF(params=columns)
    data=query.table
     
    with open('pulsars.table', 'wb') as pulsarDoc:
        pickle.dump(data, pulsarDoc)
else:
    with open('pulsars.table', 'rb') as pulsarDoc:
        data = pickle.load(pulsarDoc)
 
 
# it's faster to make a coordinate object out of floats than strings
coords=SkyCoord(data['RAJD'],data['DECJD'],unit=('deg','deg'))
 
# let's do some selections on the coordinates and catalog parameters
# we need the cos(Dec) term in here because an error in RA (or any longitude)
# changes near the poles
position_uncertainty = np.sqrt((data['RAJD_ERR'].quantity*np.cos(data['DECJD'].quantity))**2 + (data['DECJD_ERR'].quantity)**2)
 
# it's probably not a bad idea to make a histogram of these
plt.clf()
plt.hist(np.log10(position_uncertainty.to(u.arcsec).value), bins=30, cumulative=True)
plt.xlabel('log10(Position uncertainty [arcsec])')
plt.ylabel('Number of pulsars < uncertainty')
plt.plot(np.log10(max_position_uncertainty.to(u.arcsec).value)*np.array([1,1]),
          plt.gca().get_ylim(),'r--')
plt.savefig('position_uncertainty.png')
 
# make sure it's not in a globular cluster
globular_cluster=np.array(['GC' in x for x in data['ASSOC']])
# by using astropy units,  we can be assured that the position_uncertainty (in deg) and the
# max_position_uncertainty (in arcsec) are handled correctly
good = (position_uncertainty < max_position_uncertainty) & (~np.isnan(position_uncertainty)) & (~globular_cluster)
 
print('Total number of pulsars: %d' % len(data))
print('Number not in globular cluster and with position uncertainty < %.1f arcsec: %d' % (
    max_position_uncertainty.to(u.arcsec).value,
    good.sum()))
 
# now select the ones we like
data=data[good]
coords=coords[good]
position_uncertainty=position_uncertainty[good]
 
# read in RACS
racs_data=Table.read('RACS_StokesV_islands.hdf5')
     
racs_coords=SkyCoord(racs_data['ra_deg_cont'],racs_data['dec_deg_cont'],unit=('deg','deg'))
imatch, d2, _ = coords.match_to_catalog_sky(racs_coords)
# it's probably not a bad idea to make a histogram of the separations
# aa cumulative histogram will enable us to judge how many are on each side of the line
# and we will plot log(separation)
# since we want to see a small number total, use log=True for the y-axis too
plt.clf()
plt.hist(np.log10(d2.arcsec), bins=30, cumulative=True, log=True)
plt.xlabel('log10(Separation from pulsar to RACS source [arcsec])')
plt.ylabel('Number of pulsars < separation')
plt.plot(np.log10(max_separation.to(u.arcsec).value)*np.array([1,1]),
         plt.gca().get_ylim(),'r--')
plt.savefig('separation.png')
 
print('Number with separation < %.1f arcsec: %d' % (
    max_separation.to(u.arcsec).value, (d2<max_separation).sum()))
for i in range(len(data)):
    if d2[i] < max_separation:
        print('PSR %s matches island %s: flux_peak=%.1f mJy, separation=%.1f arcsec' % (
            data[i]['NAME'],
            racs_data[imatch[i]]['island_name'],
            racs_data[imatch[i]]['flux_peak'],
            d2[i].arcsec))


matches_data = [racs_data[imatch[d2 < max_separation]], data[d2 < max_separation]]
hasS400 = matches_data[1]["S400"] != np.ma.core.MaskedConstant
hasS1400 = matches_data[1]["S1400"] != np.ma.core.MaskedConstant


# Assigns values to S900 through linear scaling if either S400 or S1400 exist, if both do it uses logarithmic
S900 = np.zeros(len(hasS400))
for i in range(len(S900)):
    if hasS400[i] and hasS1400[i]:
        fit = np.polyfit(np.log([400, 1400]),np.log([matches_data[1]["S400"][i], matches_data[1]["S1400"][i]]),1)
        S900[i] = np.exp(np.polyval(fit,np.log(900)))

    elif hasS400[i]:
        S900[i] = matches_data[1]["S400"][i] * (900.0/400) ** -1.5 
    elif hasS1400[i]:
        S900[i] = matches_data[1]["S1400"][i] * (900.0/1400) ** -1.5 
    else:
        pass

matches_data[1].add_column(Column(S900,name="S900Calc"))
with open('pulsarMatches.table', 'wb') as pulsarMatchesDoc:

    pickle.dump(matches_data, pulsarMatchesDoc)
```

```python
# makes a graph of the flux density to show that they are calculated correctly
for i in range(len(matches_data[1])):
    plt.ylabel('Flux Density at x MHz [mJy]')
    plt.xlabel('Frequency [MHz]')
    plt.title("Log Graph of Flux Density vs Frequency for: " + matches_data[1]["NAME"][i])
    plt.loglog([400,900,1400], [matches_data[1]["S400"][i],matches_data[1]["S900Calc"][i],matches_data[1]["S1400"][i]], "o")
    plt.savefig('fluxDensityGraphs/' + matches_data[1]["NAME"][i] + ".png")

    plt.show()    
```
